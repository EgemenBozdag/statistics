```
dist = np.random.normal (10,2,1000)
plt.figure(figsize=(8,4), dpi = 100)
plt.hist(dist, bins = 100)
plt.show()
```

```
normal_dist =  np.random.normal(np.mean(dist),
                                         np.std(dist), 1000)

dist.sort()
normal_dist.sort()

plt.figure(figsize=(6,4), dpi = 200)
title_font = {'family': 'arial', 'color': 'darkgreen','weight': 'bold','size': 15 }
axis_font  = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 12 }
plt.title('Sample vs Normal', fontdict = title_font)
plt.xlabel('Sample', fontdict = axis_font)
plt.ylabel('Normal', fontdict = axis_font)
plt.scatter(dist, normal_dist)
plt.plot([np.min(normal_dist), np.max(normal_dist)],
         [np.min(dist), np.max(dist)], c='red')
plt.show()
```
